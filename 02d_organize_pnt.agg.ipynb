```python
import os
os.environ['USE_PYGEOS'] = '0'
import geopandas as gpd
import pandas as pd
```

## polygons

```python
es10 = gpd.read_file('./material/eg.es_agg_10km.gpkg')
de10 = gpd.read_file('./material/eg.de_agg_10km.gpkg')
es50 = gpd.read_file('./material/eg.es_agg_50km.gpkg')
de50 = gpd.read_file('./material/eg.de_agg_50km.gpkg')
es150 = gpd.read_file('./material/eg.es_agg_150km.gpkg')
de150 = gpd.read_file('./material/eg.de_agg_150km.gpkg')

es10['nuts0'] = 'ES'
de10['nuts0'] = 'DE'
es50['nuts0'] = 'ES'
de50['nuts0'] = 'DE'
es150['nuts0'] = 'ES'
de150['nuts0'] = 'DE'

es10['extent'] = 10
de10['extent'] = 10
es50['extent'] = 50
de50['extent'] = 50
es150['extent'] = 150
de150['extent'] = 150

poly10 = pd.concat([es10,de10]).reset_index(drop=True)
poly10 = poly10.drop(columns=['left','top','right','bottom'])
poly10['id'] = poly10.index
poly10.to_file('./material/polygon_agg_10km.gpkg')
poly50 = pd.concat([es50,de50]).reset_index(drop=True)
poly50 = poly50.drop(columns=['left','top','right','bottom'])
poly50['id'] = poly50.index
poly50.to_file('./material/polygon_agg_50km.gpkg')
poly150 = pd.concat([es150,de150]).reset_index(drop=True)
poly150 = poly150.drop(columns=['left','top','right','bottom'])
poly150['id'] = poly150.index
poly150.to_file('./material/polygon_agg_150km.gpkg')

# pnt = pnt.to_crs("EPSG:4326")
```

## real samples falling in this region

```python
es = gpd.read_file('./material/eg_agg_10km_es.gpkg')
de = gpd.read_file('./material/eg_agg_10km_de.gpkg')

es['id'] = es.index
de['id'] = de.index
es['nuts0'] = 'ES'
de['nuts0'] = 'DE'

ag = pd.concat([es, de])
# ag.to_crs("EPSG:4326")
```

```python
# joint with sampled data
from shapely.geometry import Point
df = pd.read_parquet('./material/soc.topsoil_organized.pq')
geometry = [Point(xy) for xy in zip(df['lon'], df['lat'])]
df = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")
df = df.to_crs(ag.crs)
```

```python
ag['polygon_id'] = ag['nuts0'].astype(str) + '-' + ag['id'].astype(str)
sdf = gpd.sjoin(df, ag, how="inner", predicate="within")
```
