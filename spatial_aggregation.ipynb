```python
import os
os.environ['USE_PYGEOS'] = '0'
import geopandas as gpd
gdf = gpd.read_file('/home/xuemeng/work_xuemeng/ai4sh_data.harmo/raw_data/EU/EU_nuts/NUTS_RG_20M_2021_3035.shp')
import pandas as pd
df = pd.read_parquet('./material/soil_organized.pq')
df = df.loc[df['soc_qa']>2]
df = df.dropna(subset=['soc'])
```

```python
from shapely.geometry import Point

geometry = [Point(xy) for xy in zip(df['lon'], df['lat'])]
points_gdf = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")  
gdf = gdf.to_crs(epsg=4326)

joined = gpd.sjoin(gdf, points_gdf, how="left", predicate="contains")
gdf["num_points"] = joined.groupby(joined.index).size()
gdf["num_points"] = gdf["num_points"].fillna(0).astype(int)

gdf = gdf.to_crs(epsg=3035)  
gdf["area_km2"] = gdf.geometry.area/1000000 
gdf['pnt_density'] = gdf['num_points']/gdf['area_km2']

gdf = gdf.sort_values(by='pnt_density',ascending=False)
```
