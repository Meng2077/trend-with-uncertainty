## read in data

```python
import pandas as pd
from sklearn.model_selection import GroupKFold,cross_val_predict
from skmap.misc import find_files, GoogleSheet, ttprint
import joblib
import numpy as np

prop = 'soc'
tgt = 'soc_log1p'
version = 'v20250301'

covs = pd.read_csv(f'./metric/feature_selected_soc_{version}.txt', header=None)[0].tolist() 
model = joblib.load(f'./model/model_rf.soc_ccc_{version}.joblib')
model.n_jobs = 30

train = pd.read_parquet(f'./material/data_train_{version}.pq')
train = train.dropna(subset=covs+[tgt],how='any')
train = train.reset_index(drop=True)

test = pd.read_parquet(f'./material/data_test_{version}.pq')
test = test.dropna(subset=covs+[tgt],how='any')
test = test.reset_index(drop=True)

# train = train.loc[train['soc']<50]
# test = test.loc[test['soc']<50]
```

## train the model

```python
# train the model
import copy
from trees_rf import cast_tree_rf, cast_node_rf, pad_leaf_outputs_to_array
import warnings
warnings.filterwarnings("ignore", message="X has feature names, but")

ttprint(f'start training')
model.fit(train[covs], train[tgt])
model_copy = copy.deepcopy(model)
ttprint('finish training')

ttprint('casting node model')
modeln = cast_node_rf(model, train[covs], train[tgt])

ttprint('casting tree model')
modelt = cast_tree_rf(model_copy)

ttprint('finish model casting')
```

## predictions

```python
dff = test[covs]#.sample(100)

ttprint('start predicting')
leaf_outputs = modeln.predict(dff)
nodes = pad_leaf_outputs_to_array(leaf_outputs, pad_value=np.nan)
nodes = nodes.T
ttprint('finish node predictions, ', nodes.shape)

trees = modelt.predict(dff)
ttprint('finish tree predictions, ', trees.shape)

# # variable importance
# importance_df = pd.DataFrame({'feature': covs, 'importance': model.feature_importances_})
# importance_df = importance_df.sort_values(by='importance', ascending=False)
# importance_df.to_csv(f'./metric/feature_importance.{prop}_{version}.csv', index=False)
```

```python
t_pred = np.nanmean(trees, axis=0)
t_pred = np.expm1(t_pred)
trees = np.expm1(trees)
t_std = np.nanstd(trees, axis=0)
```

```python
n_pred = np.nanmean(nodes, axis=0)
n_pred = np.expm1(n_pred)
nodes = np.expm1(nodes)
n_std = np.nanstd(nodes, axis=0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming n_std and t_std are already defined
# Fit a linear regression (1st degree polynomial)
slope, intercept = np.polyfit(n_std, t_std, 1)

# Create the plot
show_range = [0, 240]
fig, ax = plt.subplots(figsize=(6, 5))
hb = ax.hexbin(n_std, t_std, extent=[show_range[0], show_range[1], show_range[0], show_range[1]],
               gridsize=(30, 30), cmap='plasma_r', mincnt=1, bins='log')

ax.set_xlabel('std_nodes')
ax.set_ylabel('std_trees')
ax.plot(show_range, show_range, "-k", alpha=0.5)

# Plot the fitted line
x_vals = np.array(show_range)
y_vals = intercept + slope * x_vals
ax.plot(x_vals, y_vals, '--', label=f'Fit: slope={slope:.2f}', color='green')

# Add text annotation
ax.text(0.05, 0.95, f'Slope = {slope:.2f}', transform=ax.transAxes,
        fontsize=12, verticalalignment='top', color='green')

# Colorbar
cb = fig.colorbar(hb, shrink=0.6)
cb.set_label('Count')

plt.show()

```
